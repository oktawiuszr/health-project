File import

```python
import pandas as pd
import os

data_path="./../data"

top30data=[
    file for file in os.listdir(data_path)
    if file.endswith(".csv") and file.startswith("top")
]
low30data=[
    file for file in os.listdir(data_path)
    if file.endswith(".csv") and file.startswith("low")
]


df=pd.DataFrame(columns=["Country","Min","Max","Mean","Count","Disorder","Ranking"])
df.to_csv("./../results/exstremes_all.csv",index=False,header=True)
print(df)

for topill in top30data:
   dis_data=pd.read_csv(data_path+"/"+topill)
   illness_name=dis_data.columns[2]
   countries=dis_data["Entity"].unique()
   #print(illness_name,"\n Top countries:", str(", ".join(countries)))
   scope=dis_data.groupby("Entity")[illness_name].agg(["min","max","mean","count"]).sort_values("max",ascending=False)
   scope["Disease"]=illness_name
   scope["Ranking"]="Top"
   print(scope.head(10),"\n")
   
   scope.to_csv("./../results/exstremes_all.csv", mode="a", header=False, index=True)

for lowill in low30data:
   dis_data=pd.read_csv(data_path+"/"+lowill)
   illness_name=dis_data.columns[2]
   countries=dis_data["Entity"].unique()
   #print(illness_name,"\n Lowest countires:", str(", ".join(countries)))
   scope=dis_data.groupby("Entity")[illness_name].agg(["min","max","mean","count"]).sort_values("min", ascending=True)
   scope["Disease"]=illness_name
   scope["Ranking"]="Low"

   print(scope.head(10),"\n")
   scope.to_csv("./../results/exstremes_all.csv", mode="a", header=False, index=True)





```
