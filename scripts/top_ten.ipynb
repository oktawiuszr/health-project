```python
import pandas as pd

countries=pd.read_csv("./../data/1b-countries.csv")

for cols in countries.columns[2:]:
    base=countries[["Entity","Year",cols]]
    top_rows = base.sort_values(cols, ascending=False).head(50)#It is ok?
    low_rows = base.sort_values(cols, ascending=True).head(50)#It is ok?
    top_rows.to_csv("./../data/toprows"+cols+".csv", index=False)
    low_rows.to_csv("./../data/lowrows"+cols+".csv", index=False)
    print(cols, "\n", top_rows,"\n",low_rows)

```
